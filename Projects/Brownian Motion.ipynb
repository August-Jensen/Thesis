```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
initial_population_size = 10000
attribute_effect_on_offspring = 0.02  # 20% increase in probability of having offspring
proportion_with_attribute_initially = 0.01  # 10% of the initial population has the attribute
generations = 100
offspring_per_individual = 2  # Average number of offspring per individual

# Initialize population
population = np.zeros(initial_population_size)
# Assign the attribute to a subset of the initial population
population[:int(proportion_with_attribute_initially * initial_population_size)] = 1

# Record-keeping
attribute_counts = []

for generation in range(generations):
    # Simulate reproduction
    offspring_counts = np.random.poisson(offspring_per_individual * (1 + attribute_effect_on_offspring * population), initial_population_size)
    
    # Create next generation
    next_generation = np.repeat(population, offspring_counts)
    next_generation_size = len(next_generation)
    
    # Select individuals for the next generation (if needed)
    if next_generation_size > initial_population_size:
        next_generation = np.random.choice(next_generation, initial_population_size, replace=False)
    
    population = next_generation
    
    # Record the count of individuals with the attribute
    attribute_counts.append(np.sum(population))

# Plotting
plt.fill_between(range(generations), attribute_counts, color="skyblue", alpha=0.4)
plt.plot(attribute_counts, color="Slateblue", alpha=0.6, linewidth=2)
plt.title('Number of Individuals with Beneficial Attribute Over Generations')
plt.xlabel('Generation')
plt.ylabel('Number of Individuals with Attribute')
plt.show()
```
