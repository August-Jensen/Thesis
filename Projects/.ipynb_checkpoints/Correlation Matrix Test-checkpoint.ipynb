```python
import numpy as np

# Parameters
T = 1000  # Time steps
N = 2  # Number of assets
K = 3  # Number of factors

# Markov chain parameters
states = [0, 1]  # Two states
transition_probs = np.array([[0.95, 0.05],  # Probabilities of staying in state 0 or moving to 1
                             [0.05, 0.95]])  # Probabilities of moving to state 0 or staying in 1

# Initialize u_hat with zeros
u_hat = np.zeros((N, T))

# Initial state
current_state = np.random.choice(states, p=[0.5, 0.5])

# Simulate Markov process for u_hat
for t in range(T):
    # Based on current state, decide the next state
    current_state = np.random.choice(states, p=transition_probs[current_state])
    # Assign the state probability to u_hat (for simplicity, we just fill u_hat with state info)
    if current_state == 0:
        u_hat[:, t] = [1, 0]  # State 0 active
    else:
        u_hat[:, t] = [0, 1]  # State 1 active


# Create correlation matrices for each state
# State 0: Lower correlations
corr_matrix_state_0 = np.array([[1, 0.1, -0.4],
                                [0.1, 1, 0.3],
                                [-0.4, 0.3, 1]])

# State 1: Higher correlations
corr_matrix_state_1 = np.array([[1, 0.7, 0.6],
                                [0.7, 1, 0.8],
                                [0.6, 0.8, 1]])

# Generate residuals based on state
residuals = np.zeros((K, T))
for t in range(T):
    state = int(u_hat[0, t])  # Get the state (either 0 or 1)
    # Choose the correlation matrix based on the state
    corr_matrix = corr_matrix_state_0 if state == 0 else corr_matrix_state_1
    # Apply Cholesky decomposition to generate residuals for the current time step
    L = np.linalg.cholesky(corr_matrix)
    # Generate standard normal random variables
    z = np.random.normal(0, 1, K)
    # Compute the residuals for this time step
    residuals[:, t] = np.dot(L, z)

residuals.shape, residuals[:, :5]  # Show the shape and first 5 columns of residuals

# Start timing
start_time = time.time()

corr_matrix = np.sum(np.einsum('it,jt,nt->nijt', residuals, residuals,u_hat), axis=-1)
# Sum across T for u_hat to get the total probability weight for each state
u_hat_summed = np.sum(u_hat, axis=-1)

# Normalize each n KxK matrix in corr_matrix by the corresponding sum of u_hat
# Ensure u_hat_summed is properly shaped for broadcasting
u_hat_summed_reshaped = u_hat_summed[:, np.newaxis, np.newaxis]


corr = corr_matrix / u_hat_summed_reshaped
# End timing
end_time = time.time()
# Calculate elapsed time
elapsed_time = end_time - start_time

# Print elapsed time with 5 digits after the decimal point
print(f"Elapsed time for creating correlation matrices: {elapsed_time:.5f} seconds")

corr.shape, corr
np_corr = np.zeros((N,K,K))
for n in range(N):
    psd_1 = check_psd(corr[n])
    print(f'psd first matrix:   {psd_1}')
    np_corr[n,:,:] = scale_cholesky(corr[n,:,:])
    print(f'The {n}\'th matrix:\n{np_corr[n,:,:]}')

```
