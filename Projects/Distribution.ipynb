```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import t, norm

# Define the t distribution parameters
df = 8 # degrees of freedom

# Define the range for the x-axis
x = np.linspace(-5, 5, 1000)

# Define the PDF of the t distribution
t_pdf = t.pdf(x, df)
alpha=0
# Define the PDFs for the two normal distributions
norm1_pdf = norm.pdf(x, 0, np.sqrt(1-alpha))
norm2_pdf = norm.pdf(x, 0, np.sqrt(1+alpha))

# Calculate the sum of the two normal PDFs, weighted by 1/2
mixed_pdf = 0.5 * norm1_pdf + 0.5 * norm2_pdf

# Plot the t distribution PDF
plt.figure(figsize=(10, 6))
plt.plot(x, t_pdf, label=f't-distribution (df={df})')

# Plot the mixed normal PDF
plt.plot(x, mixed_pdf, label='Mixed Normal Distributions', linestyle='--')

plt.title('PDF of t-distribution and Mixed Normal Distributions')
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True)
plt.show()
```
