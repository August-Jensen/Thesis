```python
# Parameters
start=time.time()
K = 10000 # number of series
T = 100000 # number of observations
phi = np.random.rand(K, 1) # autoregressive parameter for each series
mu = np.random.rand(K, 1) # intercept for each series
Z = np.random.randn(K, T) # shocks for each series

# Initialize X with zeros; we'll update this with simulated values
X = np.zeros((K, T))

# Set initial values for X[:, 0] if needed, can be zeros or another specific value
X[:, 0] = np.random.randn(K)
mid = time.time()

# Simulate AR(1) process
for t in range(1, T):
    X[:, t] = mu.flatten() + np.einsum('ki,ki->k', phi, X[:, t-1].reshape(K, 1)).flatten() + Z[:, t]
end = time.time()
print(mid-start)
print(end-mid)
```

```python

from importlib import reload
reload(Final)
reload(Tools)
from Tools import RSDC
from Final import RSDCEM

# reload(EM)
# from EM import Base
O = 1000
K = 2
N = 2

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
n_iterations=10000
plot_every=25
    # Placeholder for your simulation and estimation functions
    # Since I can't run your specific RSDC and RSDCEM classes, I'm using placeholders
r =np.array((-0.5, 0.6))
def simulate_RSDC():
    # Simulate the RSDC model
    sim = RSDC(n_states=N, K_series = K, num_obs = O, rho=r transition_diagonal=0.99, deterministic=True) #sigmas = [[0,0]]
    simulated_data = sim.simulate()
    sim_param = np.vstack((sim.omega, sim.alpha, sim.beta)).T
    A = sim.rho_matrix[0][1][0]
    B = sim.rho_matrix[1][1][0]
    true_rho = np.array((A,B))
    return sim, simulated_data, true_rho, sim_param

def estimate_RSDCEM(simulated_data, sim_param):
    # Estimate the RSDC model using RSDCEM
    # Return estimated rho values
    model = RSDCEM(sim.data, univariate_parameters = sim_param)
    model.fit()
    return model.rho

# Initialize lists to store the differences
smallest_diffs = []
largest_diffs = []

n_iterations = 10000
plot_every = 50

# Main simulation and estimation loop
for i in range(n_iterations):
    sim, simulated_data, true_rho, sim_param = simulate_RSDC()
    estimated_rho = estimate_RSDCEM(simulated_data, sim_param)
    # Assuming estimated_rho is sorted or identified correctly:
    smallest_estimated_rho = min(estimated_rho) # Smallest estimated rho
    largest_estimated_rho = max(estimated_rho) # Largest estimated rho
    
    # Subtracting -0.5 from the smallest estimated rho
    smallest_diff = smallest_estimated_rho - (-0.5)
    
    # Subtracting 0.6 from the largest estimated rho
    largest_diff = largest_estimated_rho - 0.6
    # Append differences to lists
    smallest_diffs.append(smallest_diff)
    largest_diffs.append(largest_diff)
    
    # Plot histograms every plot_every iterations
    if (i + 1) % plot_every == 0:
        # Setup figure with 1 row and 3 columns: two smaller and one double-wide
        fig, axs = plt.subplots(1, 3, figsize=(15, 4), gridspec_kw={'width_ratios': [1, 1, 2]})

        # Plot histogram of smallest differences in the first subplot
        axs[0].hist(smallest_diffs, bins=30, color='blue', alpha=0.7)
        axs[0].set_title('Smallest Differences')

        # Plot histogram of largest differences in the second subplot
        axs[1].hist(largest_diffs, bins=30, color='red', alpha=0.7)
        axs[1].set_title('Largest Differences')

        # Combine smallest and largest differences for the third, double-wide plot
        total_diffs = smallest_diffs + largest_diffs
        axs[2].hist(total_diffs, bins=50, color='green', alpha=0.7)
        axs[2].set_title('Total Differences')

        # Adjust layout for clear display
        plt.tight_layout()
        plt.savefig(f'{i + 1} iteration, Convergence of RSDC.png')
        plt.close()

# Create a DataFrame to store and return the differences
diffs_df = pd.DataFrame({
    'Smallest Differences': smallest_diffs,
    'Largest Differences': largest_diffs
})
```
