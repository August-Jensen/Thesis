```python
import numpy as np
import matplotlib.pyplot as plt

# Define the parameters
gamma_values = np.linspace(0, 5, 100)  # Range of gamma values from 0 to 5
p_values = [0.25, 0.5, 0.75]  # Probabilities of success
I = 0.5# Investment amount
R = 2  # Return on investment
W0 = 2  # Initial wealth

# CRRA Utility function
def CRRA_utility(W, gamma):
    if gamma == 1:
        return np.log(W)
    else:
        return (W**(1 - gamma)) / (1 - gamma)

# Function to calculate expected utility
def expected_utility(gamma, p, W0, I, R):
    W_s = W0 - I + R  # Wealth if investment succeeds
    W_f = W0 - I  # Wealth if investment fails
    return p * CRRA_utility(W_s, gamma) + (1 - p) * CRRA_utility(W_f, gamma)

# Plotting the expected utilities for different values of gamma and p
plt.figure(figsize=(12, 8))
for p in p_values:
    utilities = [expected_utility(gamma, p, W0, I, R) for gamma in gamma_values]
    plt.plot(gamma_values, utilities, label=f'p = {p}')

plt.title("Expected Utility for Different Levels of Risk Aversion (Gamma) and Success Probabilities")
plt.xlabel("Gamma (Risk Aversion Parameter)")
plt.ylabel("Expected Utility")
plt.legend()
plt.grid(True)
plt.show()
```
