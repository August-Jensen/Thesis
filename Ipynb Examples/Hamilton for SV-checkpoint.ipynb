```python
import numpy as np
import pandas as pd

def create_gaussian_process_data(n_obs, n_states, min_sigma_diff=0.2, min_stay_prob=0.9):
    """
    Create a data series for a Gaussian process with specified number of observations,
    number of states, minimum difference in standard deviations between states,
    and minimum probability of staying in the same state.

    Parameters:
    n_obs (int): Number of observations.
    n_states (int): Number of states.
    min_sigma_diff (float): Minimum standard deviation difference between states.
    min_stay_prob (float): Minimum probability of staying in the same state.

    Returns:
    pd.Series: A data series representing the Gaussian process.
    """
    # Generate random sigmas for each state, ensuring they differ by at least min_sigma_diff
    sigmas = np.linspace(min_sigma_diff, min_sigma_diff * n_states, n_states)
    sigmas = [0.1,2,]
    # Generate transition probabilities
    transition_probs = np.zeros((n_states, n_states))
    for i in range(n_states):
        # Ensure high probability of staying in the same state
        transition_probs[i, i] = np.random.uniform(min_stay_prob, 1)
        # Assign remaining probability to other states
        remaining_prob = 1 - transition_probs[i, i]
        other_states = list(range(n_states))
        other_states.remove(i)
        for j in other_states:
            transition_probs[i, j] = remaining_prob / len(other_states)

    # Normalize rows to sum to 1 (if necessary due to rounding)
    transition_probs = transition_probs / transition_probs.sum(axis=1, keepdims=True)
    transition_probs = [[0.99, 0.01],[0.02,0.98]]
    # Generate data
    data = []
    current_state = np.random.choice(n_states)
    for _ in range(n_obs):
        # Draw from the Gaussian distribution of the current state
        data.append(np.random.normal(0, sigmas[current_state]))
        # Transition to the next state
        current_state = np.random.choice(n_states, p=transition_probs[current_state])

    return pd.Series(data)

# Example usage
n_obs = 1000
n_states = 2
y = create_gaussian_process_data(n_obs, n_states)
y.head()
```
