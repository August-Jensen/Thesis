```python
import numpy as np
import pandas as pd

# Define the number of observations
n_observations = 1000

# Transition probabilities and initial state distribution
transition_probs = np.array([[0.98, 0.01, 0.01], [0.01, 0.98, 0.01], [0.01, 0.01, 0.98]])
initial_probs = np.array([1/3, 1/3, 1/3])

# Parameters for the normal distributions for each state
mus = np.array([0, 0, 0])
sigmas = np.array([0.1, 0.7, 2])

# Initialize state and data arrays
states = np.zeros(n_observations, dtype=int)
data = np.zeros(n_observations)

# Choose initial state
states[0] = np.random.choice([0, 1, 2], p=initial_probs)

# Generate data
for i in range(1, n_observations):
    # Transition to next state based on current state's transition probabilities
    states[i] = np.random.choice([0, 1, 2], p=transition_probs[states[i-1]])

    # Generate observation from the normal distribution of the current state
    data[i] = np.random.normal(mus[states[i]], sigmas[states[i]])

# Create a DataFrame
df = pd.DataFrame({'X': data})
print(df)
import pandas as pd
import matplotlib.pyplot as plt

# Plotting the dataframe
plt.figure(figsize=(10, 6))
plt.plot(df, marker='o')
plt.title('Plot of DataFrame Values')
plt.xlabel('Index')
plt.ylabel('Values')
plt.grid(True)
plt.show()

```
