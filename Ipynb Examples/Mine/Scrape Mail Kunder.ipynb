# Til at scrape Kunder i indbakken
1. Åben Mapper der skal bruges (2. Rådgnivning & 4. Butler)
2. Åben Inspector i Mail
3. Vælg \<html> tag, og kopier 'inner'.
4. Indsæt i Dokumentet 'Mail.html' og gem.
5. Kør Block
6. Kopier output til text fil
7. Slet Unødvendigt.

```python
from bs4 import BeautifulSoup

def extract_and_replace_text(file_path, class_name):
    # Open the HTML file and parse it
    with open(file_path, 'r', encoding='utf-8') as file:
        soup = BeautifulSoup(file, 'html.parser')

    # Find the element with the specified class
    element = soup.find('div', class_=class_name)
    if not element:
        return 'Element not found'

    # Get the text and replace symbols
    text = element.get_text(strip=True)
    text = text.replace('', '\n').replace('', '\n \t')
    return text


# Example usage
file_path = 'Mail.html'  # The path to your HTML file
class_name = 'xKrjQ'  # The class name to look for


print(extract_and_replace_text(file_path, class_name))

```
