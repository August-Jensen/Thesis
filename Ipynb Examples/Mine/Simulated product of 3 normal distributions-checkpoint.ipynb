```python
# Adjusting the script to create 10 random walks
import numpy as np
import matplotlib.pyplot as plt
# Parameters
num_random_walks = 200  # Number of random walks
n = 1000  # Number of observations for each walk

# Generating the random walks
random_walks = np.zeros((num_random_walks, n))
for i in range(num_random_walks):
    random_walks[i] = np.cumsum(np.random.normal(size=n))

# Multiplying the observations at each time point
product_of_walks = np.prod(random_walks, axis=0)
# Plotting all 10 random walks

plt.figure(figsize=(12, 6))

# Looping through each random walk and plotting
for i in range(num_random_walks):
    plt.plot(random_walks[i], label=f'Random Walk {i+1}')

plt.title('10 Random Walks')
plt.xlabel('Time Step')
plt.ylabel('Accumulated Value')
plt.legend()
plt.show()


# Plotting the result
plt.figure(figsize=(12, 6))
plt.plot(product_of_walks)
plt.title('Product of 10 Random Walks Over Time')
plt.xlabel('Time Step')
plt.ylabel('Product Value')
plt.show()
```
