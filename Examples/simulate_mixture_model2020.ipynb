```python
#Importing packages we need
import numpy as np
import matplotlib.pyplot as plt #pyplot is used to plot the data
```

```python
T=250 #sample length
h1 = 4.0 #Std.dev. for state 0: "high volatility".
h2 = 0.1 #Std.dev. for state 1: "low volatility".
p = 0.8 # Probability for state 1: P(s_t = 1) = p.

#allocate vectors
y=np.zeros(T) # // observed variable
s=np.zeros(T) # // unobserved state variables (s_t = 1,2)
#simulate innovations
z=np.random.normal(loc=0, scale=1, size=T)
for t in range (T):
    
    #// Random draw from a uniform distribution on [0, 1].
    u=np.random.uniform()
    
    #// Use the uniform realization to generate hidden state s_t.
    if(u < p):
        s[t]=1 # s_t = 1, i.e. "high volatility".
    else:
        s[t]=2 # s_t = 2, i.e. "low volatility".
    
    #// Use the state s_t to generate observation y_t.
    if(s[t] == 1):
        y[t] = h1*z[t]
    else:
        y[t] = h2*z[t]
# Graph simulated observations and hidden states.
t=np.arange(0,T)
fig, axs = plt.subplots(2)
axs[0].plot(t, y, color='b')
axs[0].axhline(linewidth=1, color='k')
axs[0].set_title('log returns $x_t$')

axs[1].plot(t, s)
axs[1].set_title('State variable $s_t$')
fig.tight_layout() # Or equivalently,  "plt.tight_layout()"
plt.show()
```
