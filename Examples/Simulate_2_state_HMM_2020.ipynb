```python
#Importing packages we need
import numpy as np
import matplotlib.pyplot as plt #pyplot is used to plot the data
```

```python
#define variables
h1=4
h2=0.1
T=1000

#Note: We consider s_t = 0 and = 1 instead of = 1 and = 2. This makes the code more simple.
p11 = 0.95  #P(s_t = 1 | s_t-1 = 1). Actually, P(s_t = 0 | s_t-1 = 0).
p22 = 0.90 #P(s_t = 2 | s_t-1 = 2). Actually, P(s_t = 1 | s_t-1 = 1).

S = np.zeros(T) #state
y = np.zeros(T) #observation

#Initial value for state sequence, S_0:
S[0]   = 1 #One could also draw S[0]
y[0]   = (h1*(S[0]==0) + h2*(S[0]==1)) *np.random.normal()

for t in range(T):
    if S[t-1]==0:
        S[t]=1-(np.random.uniform()<p11)
    else:
        S[t]=np.random.uniform()<p22
    y[t]=(h1*(S[t]==0)+h2*(S[t]==1))*np.random.normal()
```

```python
# Graph simulated observations and hidden states.
t=np.arange(0,T)
fig, axs = plt.subplots(2, figsize=((10,10)))
axs[0].plot(t, y)
#axs[0].axhline(linewidth=1, color='k')
axs[0].set_title('log returns $x_t$')

axs[1].plot(t, S+1)
axs[1].set_title('State variable $s_t$')
fig.tight_layout() # Or equivalently,  "plt.tight_layout()"
plt.show()
```
