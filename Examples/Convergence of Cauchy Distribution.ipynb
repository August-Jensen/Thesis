```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
num_observations_per_draw = 1000
num_draws = 1000
degrees_of_freedom = 1  # Standard t-distribution

# Initialize array to store observations and averages
all_observations = []
averages = []

# Drawing observations and calculating averages
for _ in range(num_draws):
    new_observations = np.random.standard_t(degrees_of_freedom, size=num_observations_per_draw)
    all_observations.extend(new_observations)
    current_average = np.mean(all_observations)
    averages.append(current_average)

# Plotting the averages
plt.plot(averages)
plt.title("Average of Observations from t-Distribution Over Time")
plt.xlabel("Number of Draws")
plt.ylabel("Average of Observations")
plt.grid(True)
plt.show()
```

```python
# Parameters
num_observations_per_draw = 1000
num_draws = 100000


# Pre-allocate memory for all observations
all_observations = np.random.standard_t(degrees_of_freedom, size=num_observations_per_draw * num_draws)

# Calculate cumulative averages
cumulative_sums = np.cumsum(all_observations)
observation_counts = np.arange(1, num_observations_per_draw * num_draws + 1)
cumulative_averages = cumulative_sums / observation_counts

# Reshape cumulative averages to get the average after each set of 100 observations
reshaped_averages = cumulative_averages.reshape(num_draws, num_observations_per_draw)
final_averages = reshaped_averages[:, -1]  # Selecting the last element from each set

# Plotting the final averages
plt.plot(final_averages)
plt.title("Average of Observations from t-Distribution Over 100,000 Draws")
plt.xlabel("Number of Draws (in hundreds)")
plt.ylabel("Average of Observations")
plt.grid(True)
plt.show()
```
