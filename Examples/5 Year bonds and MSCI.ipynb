```python
import yfinance as yf
import pandas as pd

# Tickers
tickers = ['SWDA.L', 'TLT', 'BND']

# Dates
start_date = '2020-11-15'
end_date = '2023-11-15'

# Function to get closing prices
# Function to get adjusted closing prices
def get_adjusted_closing_prices(tickers, start_date, end_date):
    adjusted_prices = pd.DataFrame()
    for ticker in tickers:
        data = yf.download(ticker, start=start_date, end=end_date)
        print(data)
        adjusted_prices[ticker] = data['Adj Close']
    return adjusted_prices

# Fetch and display data
adjusted_closing_data = get_adjusted_closing_prices(tickers, start_date, end_date)
print(adjusted_closing_data)
```

```python
# Function to create an index
def create_index(data):
    indexed_data = data / data.iloc[0] * 100
    return indexed_data

# Apply the function to your DataFrame
indexed_data = create_index(adjusted_closing_data)
# Display the result
print(indexed_data)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import pandas as pd
indexed_data = indexed_data.reset_index()

# Assuming 'indexed_data' is your DataFrame and it includes a 'Date' column
# Ensure 'Date' is in datetime format
indexed_data['Date'] = pd.to_datetime(indexed_data['Date'])

# Set the plot size
plt.figure(figsize=(12, 6))  # You can adjust the width and height as needed

# Plotting
sns.lineplot(data=indexed_data, x='Date', y='SWDA.L')

# Set x-axis to display only the year
plt.gca().xaxis.set_major_locator(mdates.YearLocator())
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y'))

# Show the plot
plt.show()
```
